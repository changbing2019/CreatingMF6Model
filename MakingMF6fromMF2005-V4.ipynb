## Start Builing the MF6 model

```python
sim_name = "EAA_MF6"
sim_path = r'.\MF6'
exe_name = r'E:\projects\USGS modflow tools\mf6.4.1\bin\mf6.exe'
sim = flopy.mf6.MFSimulation(
    sim_name=sim_name, version="mf6", exe_name=exe_name, sim_ws=sim_path
)
```

```python
## add the time discritization
```

```python
nper = ml2005.nper
```

```python
perlens = ml2005.dis.perlen.array
nsteps = np.repeat(1,nper)
#nsteps[-40:] = 10
tsmultps = np.repeat(1.0,nper)
perioddata = [(float(perlen),int(nstp),float(tsmult)) for perlen, nstp,tsmult in zip(perlens,nsteps,tsmultps)]


```

```python
## add adaptive time step for kper = 140 through 180

ats_data =[]
dt0 = 0.1
dtmin = 0.01
dtmax = 1
dtadj = 2
dtfailadj = 5
for kper in range(140,nper):
    ats_data.append((kper,dt0,dtmin,dtmax,dtadj,dtfailadj))

ats_perioddata = {'maxats':40,'perioddata':ats_data}

```

```python
tdis = flopy.mf6.ModflowTdis(
    sim,
    pname="tdis",
    time_units="DAYS",
    start_date_time = '2000-12-31',
    nper=nper,
    ats_perioddata = ats_perioddata,
    perioddata=perioddata,
)
```

## Building the flow model

```python
model_name = "eaa_flow_model"
flow_model = flopy.mf6.ModflowGwf(
    sim, modelname=model_name,
    model_nam_file="{}.nam".format(model_name), 
    list = "{}.lst".format(model_name),
    print_input = True,
    print_flows =True,
    save_flows = True,
    newtonoptions=None,
)
```

## add the IMS package

```python
ims_package = flopy.mf6.ModflowIms(
    sim,
    #print_option ='SUMMARY',
    pname="ims",
    print_option="ALL",
    complexity="COMPLEX",
    outer_dvclose=5.0,
    outer_maximum=500,
    under_relaxation="DBD",
    #UNDER_RELAXATION_THETA = 0.8,
    #UNDER_RELAXATION_KAPPA = 0.0001,
    #UNDER_RELAXATION_GAMMA =0.0,
    #BACKTRACKING_NUMBER = 20,
    #BACKTRACKING_TOLERANCE = 1.05,
    #BACKTRACKING_REDUCTION_FACTOR=0.1,
    #BACKTRACKING_RESIDUAL_LIMIT = 0.002,
    inner_maximum=200,
    inner_dvclose=0.1,
    linear_acceleration="CG",
    preconditioner_levels=5,
    preconditioner_drop_tolerance=0.01,
    number_orthogonalizations=2,
)
```

```python
sim.register_ims_package(ims_package, [model_name])
```

## Well package

```python
cells = ml2005.bas6.ibound.array[0]
stress_period_data ={}
maxbound = 0
for kper in range(0,ml2005.nper):
    well = ml2005.wel.stress_period_data.__getitem__(kper=kper)
    period_t = flopy.mf6.ModflowGwfwel.stress_period_data.empty(
    flow_model,
    maxbound=len(well),
    boundnames=False)
    iw = 0
    for lay, row,col, q in well:
        cellid = (lay,row,col)
        if cells[row,col]==1 and q != 0.0:
            period_t[0][iw] = (cellid,q)
            iw += 1
        #else:
        #    print(f" the {kper} step: {cellid} not in the active area")
    period_data =  pd.DataFrame(period_t[0]).dropna().to_records(index=False)
    maxbound = max(maxbound,len(period_data))
    stress_period_data[kper] = period_data
```

```python
wel = flopy.mf6.ModflowGwfwel(
    flow_model,
    pname="wel",
    print_input=True,
    print_flows=True,
    auxiliary=None, auxmultname=None, boundnames=None,
    #auxiliary=[("var1", "var2", "var3")],
    maxbound=maxbound,
    stress_period_data = stress_period_data,
    #boundnames=True,
    save_flows=True,
)
```

```python
wel.plot(kper=0)
```

## Recharge package

```python
cells = ml2005.bas6.ibound.array[0]
rech_period ={}
lay=0
maxbound = 0
for kper in range(flow_model.nper):
    rch = ml2005.rch.rech.__getitem__(kper=kper)._Util2d__value
    newArr  = np.argwhere(rch > 0)
    maxbound =max(maxbound, len(newArr))
    rech_period_array=[]
    for idx in newArr:
        row,col =idx[0],idx[1]
        if cells[row,col]==1:
            rech_period_array.append(((lay,row,col),rch[row,col],1.0))
        
    rech_period[kper] =  rech_period_array  
        
```

```python
rch = flopy.mf6.ModflowGwfrch(
    flow_model,
    filename="{}.rch".format(model_name),
    pname="rch",
    fixed_cell=True,
    auxiliary="MULTIPLIER",
    auxmultname="MULTIPLIER",
    print_input=True,
    print_flows=True,
    save_flows=True,
    maxbound=20,
    stress_period_data=rech_period,
)
```
